⚫Find the outliers and replace them outliers

```python
sns.boxplot(df.Length)
```

```python
df.Length.median()
```

```python
q1=df.Length.quantile(0.25)  #(Q1)
q3=df.Length.quantile(0.75)  #(Q3)
IQR=q3-q1
lower_limit= q1 - 1.5*IQR
```

```python
df['Length']=np.where(df['Length']<lower_limit,0.545,df['Length'])
```

```python
sns.boxplot(df.Length)
```

```python
sns.boxplot(df.Diameter)
```

```python
df.Diameter.median()
```

```python
q1=df.Diameter.quantile(0.25)  #(Q1)
q3=df.Diameter.quantile(0.75)  #(Q3)
IQR=q3-q1
lower_limit= q1 - 1.5*IQR
```

```python
df['Diameter']=np.where(df['Diameter']<lower_limit,0.425,df['Diameter'])
```

```python
sns.boxplot(df.Diameter)
```

```python
sns.boxplot(df.Height)
```

```python
q1=df.Height.quantile(0.25)  #(Q1)
q3=df.Height.quantile(0.75)  #(Q3)
IQR=q3-q1
lower_limit= q1 - 1.5*IQR
upper_limit= q3 + 1.5*IQR
```

```python
df.Height.mean()
```

```python
df['Height']=np.where(df['Height']<lower_limit,0.13951639932966242,df['Height'])
df['Height']=np.where(df['Height']>upper_limit,0.13951639932966242,df['Height'])
```

```python
sns.boxplot(df.Height)
```

⚫Check for Categorical columns and perform encoding

```python
df=pd.get_dummies(df,columns=['Sex'])
df.head()
```

⚫Split the data into dependent and independent variables

```python
y=df['Length']#DEPENDANT VARIABLE
y
```

```python
#independent variable

X=df.drop(columns=['Height'],axis=1)
X.head()
```

⚫Scale the independent variables

```python
from sklearn.preprocessing import scale
```

```python
x_scaled=pd.DataFrame(scale(X),columns=X.columns)
x_scaled.head()
```

⚫Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

⚫BUILD THE MODEL

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
r=Ridge()
l=Lasso()
```

```python
r.fit(X_train,y_train)
```

```python
l.fit(X_train,y_train)
```

⚫TRAINING THE MODEL

```python
pred1=r.predict(X_test)
pred1
```

```python
pred1_train=r.predict(X_train)
```

```python
pred2=l.predict(X_test)
pred2
```

```python
pred2_train=l.predict(X_train)
```

```python
profit=pd.DataFrame({'Actual':y_test,'ridge_pred':pred1,'lasso_pred':pred2})
profit.head(10)
```

⚫TESTING THE MODEL

```python
#TESTING BY RANDOM VALUE 
p=r.predict([[-0.645490	,-0.504337	,-1.198473	,-0.641898	,-0.607685	,-0.726212,	-0.638217	,1.571544,	-0.674834	,-0.688018	,1.316677]])
print(p)
p1=l.predict([[-0.645490,-0.504337,-1.198473,-0.641898,-0.607685,-0.726212,	-0.638217,1.571544,	-0.674834,-0.688018	,1.316677]])
print(p1)
```

⚫EVALUATING THE MODEL


```python
from sklearn import metrics
```

```python
# R-Square
# testing accuracy for both model
print(metrics.r2_score(y_test,pred1))
print(metrics.r2_score(y_test,pred2))
```

```python
#Training accuracy for both model

print(metrics.r2_score(y_train,pred1_train))
print(metrics.r2_score(y_train,pred2_train))
```

```python
## MSE(Mean square error)

print(metrics.mean_squared_error(y_test,pred1))
print(metrics.mean_squared_error(y_test,pred2))
```

```python
## RMSE

print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.sqrt(metrics.mean_squared_error(y_test,pred2)))
```
